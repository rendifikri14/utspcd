```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca citra
img = cv2.imread('Panda.png', 0)

# Definisi Kernel
# LPF (Low Pass Filter) - Mengaburkan
kernel_lpf = np.ones((3,3), np.float32) / 9

# HPF (High Pass Filter) - Deteksi Tepi
kernel_hpf = np.array([[-1, -1, -1],
                       [-1,  8, -1],
                       [-1, -1, -1]])

# BPF (Band Pass Filter / Sharpening) - Penajaman
kernel_bpf = np.array([[ 0, -1,  0],
                       [-1,  5, -1],
                       [ 0, -1,  0]])

# Terapkan Filter (Konvolusi)
res_lpf = cv2.filter2D(img, -1, kernel_lpf)
res_hpf = cv2.filter2D(img, -1, kernel_hpf)
res_bpf = cv2.filter2D(img, -1, kernel_bpf)

# Hitung Rata-rata Piksel
print(f"Rata-rata Piksel Asli: {np.mean(img):.2f}")
print(f"Rata-rata Piksel LPF: {np.mean(res_lpf):.2f}")
print(f"Rata-rata Piksel HPF: {np.mean(res_hpf):.2f}")
print(f"Rata-rata Piksel BPF: {np.mean(res_bpf):.2f}")

# Tampilkan Hasil
plt.figure(figsize=(15, 5))
plt.subplot(141), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(142), plt.imshow(res_lpf, cmap='gray'), plt.title('LPF (Blur)')
plt.subplot(143), plt.imshow(res_hpf, cmap='gray'), plt.title('HPF (Edge)')
plt.subplot(144), plt.imshow(res_bpf, cmap='gray'), plt.title('BPF (Sharpen)')
plt.show()
```
