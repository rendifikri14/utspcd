```python
# @title Default title text
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Baca citra grayscale
# Ganti 'gambar_anda.jpg' dengan nama file gambar yang diupload ke Colab, termasuk ekstensi file yang benar.
# IMPORTANT: Please update the path below with the correct file path and name of your image, including the extension!
img = cv2.imread('Panda.png', 0)

# Tambahkan pengecekan untuk memastikan gambar berhasil dimuat
if img is None:
    print("Error: Image could not be loaded. Please check the file path and ensure the file exists.")
    # Exit or handle the error appropriately
    # For now, we'll exit to prevent further errors.
    exit()

# 2. Terapkan Rumus

# a. Citra Negatif
# Rumus: G = 255 - F
img_negative = 255 - img

# b. Citra Logaritmik
# Rumus: G = c * log(1 + F)
# c = 255 / log(1 + max_pixel_value)
c = 255 / np.log(1 + np.max(img))
img_log = c * (np.log(1 + img))
img_log = np.array(img_log, dtype=np.uint8) # Konversi kembali ke integer

# 3. Tampilkan Gambar dan Histogram
fig, ax = plt.subplots(3, 2, figsize=(12, 12))

# Citra Asli
ax[0, 0].imshow(img, cmap='gray')
ax[0, 0].set_title('Citra Asli')
ax[0, 1].hist(img.ravel(), 256, [0, 256])
ax[0, 1].set_title('Histogram Asli')

# Citra Negatif
ax[1, 0].imshow(img_negative, cmap='gray')
ax[1, 0].set_title('Citra Negatif')
ax[1, 1].hist(img_negative.ravel(), 256, [0, 256])
ax[1, 1].set_title('Histogram Negatif')

# Citra Logaritmik
ax[2, 0].imshow(img_log, cmap='gray')
ax[2, 0].set_title('Citra Logaritmik')
ax[2, 1].hist(img_log.ravel(), 256, [0, 256])
ax[2, 1].set_title('Histogram Logaritmik')

plt.tight_layout()
plt.show()
```
